```python
import requests
import time
from bs4 import BeautifulSoup as BS
from bs4 import Comment
```

```python
URL = r"https://www.atomuhr.de/"
```

```python
atime = ""
adate = ""

while (True):

    html = requests.get(URL).content
    soup = BS(html, 'html.parser')
    comments = soup.find_all(string=lambda text: isinstance(text, Comment))
    
    for c in comments:
        if ("@" in c):
            bdate = (c[c.find("@")+1:c.find("@")+12])
            btime = (c[c.find("@")+13:c.find("@")+21])

        if (bdate != adate):
            adate = bdate
            print(adate)

        if (btime != atime):
            atime = btime
            print(atime)

    time.sleep(5)
```
